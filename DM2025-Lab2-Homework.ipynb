This section summarizes the preprocessing, feature engineering, and model development used in the Kaggle competition.

---

### **1.1 Preprocessing Steps**

For this competition, I applied a series of preprocessing techniques designed to **preserve emotional cues** while cleaning noise in the dataset. Excessive cleaning can remove crucial features such as emoji, punctuation, and negation, so a balanced preprocessing strategy was used.

#### **Main Preprocessing Steps**
- **Lowercasing** all text for normalization.  
- **Removing URLs and user mentions**, which do not contribute to emotion detection.  
- **Expanding contractions** (e.g., `"can't" → "can not"`).  
- **Emoji mapping** to emotion tokens  
  - 😂 → `JOY_EMOJI`  
  - 😡 → `ANGER_EMOJI`  
  - 😭 → `CRY_EMOJI`  
- **Keyword injection** using custom emotion dictionaries  
  - Example: “happy”, “excited” → `JOY_WORD`  
- **Negation merging**, e.g. `"not good" → "not_good"`.  
- **Removing non-alphanumeric characters** while preserving underscore-based markers.  
- **Lemmatization** using WordNetLemmatizer.

These steps help retain expressive patterns important for emotional interpretation, such as emojis, intense keywords, punctuation, and negations.

### **1.2 Feature Engineering Steps**

A hybrid feature engineering design was used to combine **semantic meaning** and **sentiment cues**:

#### **(1) MiniLM Sentence Embeddings**
- Used **all-MiniLM-L6-v2** (384-dimensional vectors)
- Captures contextual and semantic relationships beyond word-level features

#### **(2) Numeric Emotion Features**
To enhance emotion detection, I engineered 11 additional numerical features:

- Word count  
- Number of exclamation marks  
- Number of question marks  
- Count of uppercase letters  
- Counts of injected emotion tokens (`JOY_WORD`, `ANGER_WORD`, etc.)  
- Counts of emoji emotion tokens (`JOY_EMOJI`, `CRY_EMOJI`, ...)  

These features help when semantic meaning is unclear but emotional punctuation or cues are strong.

#### **(3) Final Feature Vector Structure**

This combination captures both **semantic meaning** and **surface-level emotional expressions**.

### **1.3 Explanation of Your Model**

I trained three distinct machine-learning models using the engineered features and combined them using an ensemble.

#### **(1) Logistic Regression**

#### **(2) Linear SVC**

## 1. Preprocessing Steps

```python
### Add the code related to the preprocessing steps in cells inside this section
# ================================
# 1. Preprocessing Steps
# ================================

import os
import re
import json
import pandas as pd
from sklearn.model_selection import train_test_split

# 如果你在本機 / Colab，要記得確定路徑正確
DATA_DIR = "./kaggle_data"

# 檔案路徑
posts_path = os.path.join(DATA_DIR, "final_posts.json")
emotion_path = os.path.join(DATA_DIR, "emotion.csv")
ident_path = os.path.join(DATA_DIR, "data_identification.csv")

print("Loading data...")

# ===============================
# 正確讀取 final_posts.json（巢狀 JSON）
# ===============================
with open(posts_path, "r", encoding="utf-8") as f:
    data = json.load(f)  # data 是 list，每筆資料有一個 root 欄位

records = []

for item in data:
    try:
        post_obj = item["root"]["_source"]["post"]
        post_id = post_obj.get("post_id")
        text = post_obj.get("text")
        records.append({"id": post_id, "text": text})
    except KeyError as e:
        print("Skipping due to missing key:", e)

posts_df = pd.DataFrame(records)

print("Parsed posts_df:")
print(posts_df.head())
print("posts_df shape:", posts_df.shape)


# ===============================
# 載入其他兩個 csv
# ===============================
emotion_df = pd.read_csv(emotion_path)
print("Emotion columns:", emotion_df.columns.tolist())

ident_df = pd.read_csv(ident_path)
print("Identification columns:", ident_df.columns.tolist())


# ===============================
# 依照 id 合併
# ===============================
df = posts_df.merge(ident_df, on="id", how="inner")
df = df.merge(emotion_df, on="id", how="left")

print("Merged df shape:", df.shape)
print(df.head())

assert "text" in df.columns
assert "split" in df.columns


# ===============================
# 分 train / test
# ===============================
train_df = df[df["split"] == "train"].copy()
test_df = df[df["split"] == "test"].copy()

print("Train size:", train_df.shape)
print("Test size:", test_df.shape)

train_df["text"] = train_df["text"].fillna("")
test_df["text"] = test_df["text"].fillna("")


# ===============================
# 文字清理
# ===============================
url_pattern = re.compile(r"http\S+|www\.\S+")
user_pattern = re.compile(r"@\w+")

def clean_text(text: str) -> str:
    text = text.lower()
    text = url_pattern.sub(" ", text)
    text = user_pattern.sub(" ", text)
    text = re.sub(r"#", " ", text)
    text = re.sub(r"[^a-zA-Z0-9\s]", " ", text)
    text = re.sub(r"\s+", " ", text).strip()
    return text

print("Cleaning text...")

train_df["clean_text"] = train_df["text"].apply(clean_text)
test_df["clean_text"] = test_df["text"].apply(clean_text)

print(train_df[["text", "clean_text"]].head())
```

```python
# ================================
# TEXT PREPROCESSING (強化版)
# ================================
import re
import nltk
nltk.download("wordnet")
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()

# 🔥 1. Contraction Expansion
contractions = {
    "can't": "can not", "won't": "will not", "n't": " not",
    "i'm": "i am", "it's": "it is", "he's": "he is", "she's": "she is",
    "they're": "they are", "we're": "we are", "you're": "you are",
    "i've": "i have", "we've": "we have", "they've": "they have",
    "isn't": "is not", "aren't": "are not", "wasn't": "was not",
    "weren't": "were not"
}

def expand_contractions(text):
    for c, e in contractions.items():
        text = re.sub(rf"\b{c}\b", e, text)
    return text

# 🔥 2. Emoji → Emotion
emoji_map = {
    "😀":"joy", "😃":"joy", "😄":"joy", "😁":"joy", "😆":"joy",
    "😂":"joy", "😊":"joy", "🙂":"joy", "😍":"joy",

    "😡":"anger", "😠":"anger",

    "😢":"sadness", "😭":"sadness",

    "😱":"fear", "😨":"fear",

    "🤢":"disgust", "🤮":"disgust",

    "😮":"surprise", "😯":"surprise", "😲":"surprise"
}

def replace_emojis(text):
    for e, label in emoji_map.items():
        text = text.replace(e, f" {label} ")
    return text

# 🔥 3. Negation Handling
def join_negations(text):
    return re.sub(r"(not|no|never)\s+(\w+)", r"\1_\2", text)

# 🔥 4. Lemmatization
def lemmatize(text):
    return " ".join([lemmatizer.lemmatize(w) for w in text.split()])

# 🔥 5. full cleaning
def full_clean(text):
    text = text.lower()
    text = expand_contractions(text)
    text = replace_emojis(text)
    text = join_negations(text)
    text = re.sub(r"http\S+|www\.\S+", " ", text)
    text = re.sub(r"@\w+", " ", text)
    text = re.sub(r"#", " ", text)
    text = re.sub(r"[^a-zA-Z0-9\s_]", " ", text)
    text = re.sub(r"\s+", " ", text).strip()
    text = lemmatize(text)
    return text

print("Running enhanced preprocessing...")

train_df["clean2"] = train_df["clean_text"].apply(full_clean)
test_df["clean2"] = test_df["clean_text"].apply(full_clean)

print(train_df[["text", "clean2"]].head())
```

```python
# ================
# Preprocessing
# ================

import re
import numpy as np
import pandas as pd

# -------- Emoji mapping --------
emoji_map = {
    "😂": "LAUGH_EMOJI",
    "🤣": "LAUGH_EMOJI",
    "😭": "CRY_EMOJI",
    "😢": "CRY_EMOJI",
    "😡": "ANGRY_EMOJI",
    "🤬": "ANGRY_EMOJI",
    "😱": "FEAR_EMOJI",
    "😨": "FEAR_EMOJI",
    "🤢": "DISGUST_EMOJI",
    "🤮": "DISGUST_EMOJI",
    "😮": "SURPRISE_EMOJI",
    "😲": "SURPRISE_EMOJI",
    "😃": "JOY_EMOJI",
    "😊": "JOY_EMOJI",
    "❤️": "LOVE_EMOJI"
}

# -------- Emotion keyword dictionary --------
emotion_keywords = {
    "joy": ["happy", "joy", "love", "glad", "excited", "wonderful", "awesome", "great"],
    "anger": ["angry", "furious", "hate", "mad", "pissed", "rage"],
    "sadness": ["sad", "cry", "depressed", "lonely", "heartbroken"],
    "fear": ["fear", "scared", "afraid", "terrified", "panic"],
    "disgust": ["disgust", "gross", "ew", "nasty"],
    "surprise": ["surprise", "surprised", "omg", "wtf", "shocked"],
}

# -------- Preprocess function --------
def preprocess_text(t):
    if type(t) != str:
        return ""

    # basic clean
    t = t.lower()
    t = re.sub(r"http\S+|www\.\S+", " URL ", t)
    t = re.sub(r"@\w+", " USER ", t)

    # emoji mapping
    for e, token in emoji_map.items():
        t = t.replace(e, f" {token} ")

    # keyword inject
    for emo, words in emotion_keywords.items():
        for w in words:
            t = re.sub(rf"\b{w}\b", f" {emo.upper()}_WORD ", t)

    # remove special chars
    t = re.sub(r"[^a-zA-Z0-9_\s]", " ", t)
    t = re.sub(r"\s+", " ", t).strip()
    return t

print("Applying enhanced preprocessing...")
train_df["clean3"] = train_df["clean2"].apply(preprocess_text)
test_df["clean3"] = test_df["clean2"].apply(preprocess_text)
```

## 2. Feature Engineering Steps

```python
# ================================
# MiniLM Embeddings
# ================================
!pip install -q sentence-transformers

from sentence_transformers import SentenceTransformer
st_model = SentenceTransformer("all-MiniLM-L6-v2")

print("Encoding MiniLM...")

X_train_minilm = st_model.encode(
    train_df["clean3"].tolist(),
    batch_size=64,
    show_progress_bar=True,
    convert_to_numpy=True
)

X_test_minilm = st_model.encode(
    test_df["clean3"].tolist(),
    batch_size=64,
    show_progress_bar=True,
    convert_to_numpy=True
)

y_train = train_df["emotion"].values

print("MiniLM shape:", X_train_minilm.shape)
```

```python
def numeric_features(text):
    return [
        len(text.split()),              # length
        text.count("!"),                # exclamation
        text.count("?"),                # question
        sum(c.isupper() for c in text), # upper chars
        text.count("JOY_WORD"),
        text.count("ANGER_WORD"),
        text.count("FEAR_WORD"),
        text.count("SADNESS_WORD"),
        text.count("DISGUST_WORD"),
        text.count("SURPRISE_WORD"),
        text.count("EMOJI")
    ]

print("Building numeric features...")

numeric_train = np.array([numeric_features(t) for t in train_df["clean3"]])
numeric_test = np.array([numeric_features(t) for t in test_df["clean3"]])

print("Numeric feature shape:", numeric_train.shape)
```

```python
X_train_full = np.concatenate([X_train_minilm, numeric_train], axis=1)
X_test_full = np.concatenate([X_test_minilm, numeric_test], axis=1)

print("Final feature shape:", X_train_full.shape)
```

## 3. Model Implementation Steps

```python
import time
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.svm import LinearSVC
from sklearn.linear_model import LogisticRegression
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import f1_score


# =====================================================
# 1) Label Encoding
# =====================================================
le = LabelEncoder()
y_encoded = le.fit_transform(y_train)
print("Label classes:", le.classes_)


# =====================================================
# 2) StandardScaler（超重要）
# =====================================================
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_full)
X_test_scaled  = scaler.transform(X_test_full)


# =====================================================
# 3) Train/Validation split
# =====================================================
X_tr, X_val, y_tr, y_val = train_test_split(
    X_train_scaled, y_encoded,
    test_size=0.15,
    random_state=42,
    stratify=y_encoded
)


# =====================================================
# 4) Logistic Regression（很強）
# =====================================================
lr_clf = LogisticRegression(
    max_iter=2000,
    solver="lbfgs",
    n_jobs=-1
)
lr_clf.fit(X_tr, y_tr)
pred_lr = lr_clf.predict(X_val)
print("LR F1:", f1_score(y_val, pred_lr, average="macro"))


# =====================================================
# 5) Linear SVC（很強）
# =====================================================
svc_clf = LinearSVC()
svc_clf.fit(X_tr, y_tr)
pred_svc = svc_clf.predict(X_val)
print("SVC F1:", f1_score(y_val, pred_svc, average="macro"))


# =====================================================
# 6) 強化版 MLP（僅 5 epochs，快速＋穩定）
# =====================================================
mlp_clf = MLPClassifier(
    hidden_layer_sizes=(768, 384, 192),
    activation="relu",
    solver="adam",
    alpha=1e-5,
    learning_rate="adaptive",
    learning_rate_init=0.001,
    batch_size=128,
    max_iter=1,              # 1 epoch per fit()
    warm_start=True
)

print("\n===== Start MLP Training (5 epochs) =====")
EPOCHS = 5
best_f1 = -1
best_mlp = None

for epoch in range(1, EPOCHS + 1):
    start = time.time()
    mlp_clf.fit(X_tr, y_tr)
    pred_val = mlp_clf.predict(X_val)
    f1_macro = f1_score(y_val, pred_val, average="macro")
    end = time.time()

    print(f"Epoch {epoch}/5 | F1={f1_macro:.4f} | time={end - start:.2f} sec")

    if f1_macro > best_f1:
        best_f1 = f1_macro
        best_mlp = MLPClassifier(
            hidden_layer_sizes=mlp_clf.hidden_layer_sizes,
            activation="relu",
            solver="adam"
        )
        best_mlp.coefs_ = [w.copy() for w in mlp_clf.coefs_]
        best_mlp.intercepts_ = [b.copy() for b in mlp_clf.intercepts_]

print(f"\nBest MLP F1: {best_f1:.4f}")


# =====================================================
# 7) Final Ensemble (simple majority vote)
# =====================================================
def majority_vote(pred_arrays):
    final = []
    for col in np.vstack(pred_arrays).T:
        values, counts = np.unique(col, return_counts=True)
        final.append(values[np.argmax(counts)])
    return np.array(final)

pred_ensemble = majority_vote([pred_lr, pred_svc, pred_val])
print("\nEnsemble F1:", f1_score(y_val, pred_ensemble, average="macro"))


# =====================================================
# 8) Retrain ALL on FULL TRAINING DATA
# =====================================================
print("\n===== Training FINAL MODELS on FULL DATA =====")

lr_clf.fit(X_train_scaled, y_encoded)
svc_clf.fit(X_train_scaled, y_encoded)

mlp_final = MLPClassifier(
    hidden_layer_sizes=(768, 384, 192),
    activation="relu",
    solver="adam",
    alpha=1e-5,
    learning_rate="adaptive",
    learning_rate_init=0.001,
    batch_size=128,
    max_iter=5,          # 5 epochs full train
    warm_start=False
)

mlp_final.fit(X_train_scaled, y_encoded)


# =====================================================
# 9) Predict Test Set (Ensemble)
# =====================================================
test_pred = majority_vote([
    lr_clf.predict(X_test_scaled),
    svc_clf.predict(X_test_scaled),
    mlp_final.predict(X_test_scaled)
])

test_pred_text = le.inverse_transform(test_pred)

submission = pd.DataFrame({
    "id": test_df["id"],
    "emotion": test_pred_text
})

submission.to_csv("submission_3model_ensemble.csv", index=False)
print("\nSaved submission_3model_ensemble.csv")
```
